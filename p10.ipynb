```python
import torch
from torch import nn
```

```python
x1, x2 = torch.tensor(1., requires_grad=True), torch.tensor(2., requires_grad=True)
```

```python
num_epochs, lr = 1000, 0.01
for _ in range(num_epochs):
    y1 = x1 ** 2
    y2 = -(x2 - 4) ** 4 - 2
    d = torch.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)
    d.backward()
    x1.data -= lr * x1.grad
    x2.data -= lr * x2.grad
    x1.grad.zero_()
    x2.grad.zero_()
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
# 定义两个函数
def func1(x):
    return x**2

def func2(x):
    return -(x - 4)**4 - 2

# 设置x范围
x = np.linspace(-2, 6, 400)  # x1 和 x2 范围从 -2 到 6

# 计算对应的y值
y1 = func1(x)
y2 = func2(x)

# 创建图形和坐标轴
fig, ax = plt.subplots(figsize=(8, 6))

# 绘制两个函数
ax.plot(x, y1, label=r'$y = x^2$', color='blue')
ax.plot(x, y2, label=r'$y = -(x-4)^4 - 2$', color='red')

# 设置轴范围
ax.set_xlim(-2, 6)  # 限定x轴范围
ax.set_ylim(-40, 10)  # 限定y轴范围

# 设置标签和标题
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Graphs of $y = x^2$ and $y = -(x-4)^4 - 2$')

# 添加图例
ax.legend()

# 显示图形
plt.grid(True)
plt.show()
```
